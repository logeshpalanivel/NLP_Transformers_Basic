```python
from transformers import pipeline

c = pipeline("sentiment-analysis")
c('this is the worst day in my life')
```

```python
c(['i hate this so much','i feel very happy today'])
```

```python
c = pipeline('zero-shot-classification')
c('I Have Exams',candidate_labels=['ground','school','temple'])
```

```python
g = pipeline('text-generation')
g('Im going to college to')
```

```python
h = pipeline('text-generation',model='distilgpt2')
h('im going to scholl to play an ',max_length=30,num_return_sequences=2)
```

```python
u = pipeline('fill-mask')
u('im going to work as an <mask> in this company',top_k=2)
```

```python
ner = pipeline('ner',grouped_entities = True)
ner('My name is Logesh and I work at Infosys in Coimbatore')
```

```python
q = pipeline('question-answering')
q(
    question = 'what is my name?',
    context='My name is logesh,working in Infusys in coimbatore'
)
```
